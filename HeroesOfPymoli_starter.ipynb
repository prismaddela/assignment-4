```python
# Dependencies and Setup
import pandas as pd
import numpy as np
```

```python
# File to Load (Remember to Change These)
file_to_load = "Resources/purchase_data.csv"

# Read Purchasing File and store into Pandas data frame
purchase_data_df = pd.read_csv(file_to_load)
purchase_data_df.head()
```

## Player Count

* Display the total number of players


```python
players=len(purchase_data_df["SN"].unique())
totalplayers=pd.DataFrame({"players": [players]})
totalplayers
```

## Purchasing Analysis (Total)

* Run basic calculations to obtain number of unique items, average price, etc.


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame


```python
# Pull unique values
UniqueItems = len(purchase_data_df["Item Name"].unique())
AvgPrice = purchase_data_df["Price"].mean()
Purchases = len(purchase_data_df["Item Name"])
Revenue = purchase_data_df["Price"].sum()

# Create Purchasing DataFrame
PurchasingAnalysis = pd.DataFrame({"Unique Items": [UniqueItems],
                                           "Average Price": [AvgPrice],
                                           "Purchases": [Purchases],
                                           "Total Revenue": [Revenue]})

# DataFrame formatting
PurchasingAnalysis["Average Price"] = PurchasingAnalysis["Average Price"].map("${:.2f}".format)
PurchasingAnalysis = PurchasingAnalysis[["Unique Items", "Average Price", "Purchases"]]

PurchasingAnalysis
```

## Gender Demographics

* Percentage and Count of Male Players


* Percentage and Count of Female Players


* Percentage and Count of Other / Non-Disclosed




```python
# Data manipulation
duplicate = purchase_data_df.drop_duplicates(subset='SN', keep="first")
TotalGender = duplicate["Gender"].count()
Male = duplicate["Gender"].value_counts()['Male']
Female = duplicate["Gender"].value_counts()['Female']
NonDisclosed = TotalGender - Male - Female

MalePercent = (Male / TotalGender) * 100
FemalePercent = (Female / TotalGender) * 100
NonPercent = (NonDisclosed / TotalGender) * 100

# Create Gender DataFrame
Gender_Demo = pd.DataFrame({"": ['Male', 'Female', 'Other/Non-Disclosed'],
                            "Percentage of Players": [MalePercent, FemalePercent, NonPercent],
                            "Total Count": [Male, Female, NonDisclosed]})

# DataFrame formatting
Gender_Demo["Percentage of Players"] = Gender_Demo["Percentage of Players"].map("{:.2f}%".format)
Gender_Demo = Gender_Demo.set_index('')
Gender_Demo
```


## Purchasing Analysis (Gender)

* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. by gender




* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
# Group by Gender
gendergroup_df = purchase_data_df.groupby(["Gender"])

# Grouped Data 
PurchaseCount = gendergroup_df["SN"].count()
PurchasePrice = gendergroup_df["Price"].mean()
PurchaseTotal = gendergroup_df["Price"].sum()

# Delete all duplicates 
duplicate = purchase_data_df.drop_duplicates(subset='SN', keep="first")
grouped_dupe = duplicate.groupby(["Gender"])

# Find normalized data
PurchaseNorm = (purchase_data_df["Price"].sum() / grouped_dupe["SN"].count())

# New DataFrame
PurchaseGenderAnalysis = pd.DataFrame({"Purchase Count": PurchaseCount,
                              "Average Purchase Price": PurchasePrice,
                              "Total Purchase Value": PurchaseTotal,
                              "Normalized Totals": PurchaseNorm})

# DataFrame formatting
PurchaseGenderAnalysis["Average Purchase Price"] = PurchaseGenderAnalysis["Average Purchase Price"].map("${:.2f}".format)
PurchaseGenderAnalysis["Total Purchase Value"] = PurchaseGenderAnalysis["Total Purchase Value"].map("${:.2f}".format)
PurchaseGenderAnalysis["Normalized Totals"] = PurchaseGenderAnalysis["Normalized Totals"].map("${:.2f}".format)
PurchaseGenderAnalysis = PurchaseGenderAnalysis[["Purchase Count", "Average Purchase Price", "Total Purchase Value", "Normalized Totals"]]
PurchaseGenderAnalysis
```

## Age Demographics

* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table


```python
# Bins
bin = [0,10,15,20,25,30,35,40,45,50,150]
binLab = ['Under 10', '10 - 14', '15 - 19','20 - 24', '25 - 29','30 - 34', '35 - 39','40 - 44', '45 - 49','Over 50']

# Add bins dataframe and groupby
bins_df = purchase_data_df.copy()
bins_df["Age Groups"] = pd.cut(bins_df["Age"], bin, labels=binLab)
group_bin = bins_df.groupby(["Age Groups"])

# Data 
BinsCount = group_bin["SN"].count()
CountTotal = purchase_data_df["SN"].count()
percentage = (BinsCount / CountTotal) * 100
percentage

# Create new DF
AgePercent = pd.DataFrame({"Total Count": BinsCount,
                         "Percentage of Players": percentage})

# Format
AgePercent["Percentage of Players"] = AgePercent["Percentage of Players"].map("{:.2f}%".format)
AgePercent.head(10)
```

* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting

```python
# Binning
bin = [0,10,15,20,25,30,35,40,45,50,150]
binLab = ['Under 10', '10 - 14', '15 - 19','20 - 24', '25 - 29','30 - 34', '35 - 39','40 - 44', '45 - 49','Over 50']

# Add bins to new dataframe and groupby
binning_df = purchase_data_df.copy()
binning_df["Age Groups"] = pd.cut(binning_df["Age"], bin, labels=binLab)
binsColumn = pd.cut(binning_df["Age"], bin, labels=binLab)
grouped_bin = binning_df.groupby(["Age Groups"])

# Data 
binPCount = grouped_bin["Age"].count()
binPAverage = grouped_bin["Price"].mean()
binPTotal = grouped_bin["Price"].sum()

# Normalize data 
binningduplicate = purchase_data_df.drop_duplicates(subset='SN', keep="first")
binningduplicate["Age Groups"] = pd.cut(binningduplicate["Age"], bin, labels=binLab)
binningduplicate = binningduplicate.groupby(["Age Groups"])

binningNorm = (grouped_bin["Price"].sum() / binningduplicate["SN"].count())
binningNorm

# Create new and format
Age_Demo = pd.DataFrame({"Purchase Count": binPCount,
                         "Average Purchase Price": binPAverage,
                         "Total Purchase Value": binPTotal,
                         "Normalized Totals": binningNorm})

Age_Demo["Average Purchase Price"] = Age_Demo["Average Purchase Price"].map("${:.2f}".format)
Age_Demo["Total Purchase Value"] = Age_Demo["Total Purchase Value"].map("${:.2f}".format)
Age_Demo["Normalized Totals"] = Age_Demo["Normalized Totals"].map("${:.2f}".format)
Age_Demo = Age_Demo[["Purchase Count", "Average Purchase Price", "Total Purchase Value", "Normalized Totals"]]
Age_Demo.head(10)
```

## Top Spenders

* Run basic calculations to obtain the results in the table below


* Create a summary data frame to hold the results


* Sort the total purchase value column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
# Data 
groupBySN = purchase_data_df.groupby(["SN"])
groupSNCount = groupBySN["Item ID"].count()
groupSNTotal = groupBySN["Price"].sum()
groupSNAvg = (groupSNTotal / groupSNCount)

# Build DF and format
SN_Demo = pd.DataFrame({"Purchase Count": groupSNCount,
                         "Average Purchase Price": groupSNAvg,
                         "Total Purchase Value": groupSNTotal})

SN_Demo = SN_Demo.sort_values("Total Purchase Value", ascending=False) 
SN_Demo["Average Purchase Price"] = SN_Demo["Average Purchase Price"].map("${:.2f}".format)
SN_Demo["Total Purchase Value"] = SN_Demo["Total Purchase Value"].map("${:.2f}".format)
SN_Demo = SN_Demo[["Purchase Count", "Average Purchase Price", "Total Purchase Value"]]
SN_Demo.head()
```

```python
# Data
groupItem = purchase_data_df.groupby(["Item ID", "Item Name"])
groupItemC = groupItem["SN"].count()
groupPriceSum = groupItem["Price"].sum()
groupItemP = (groupPriceSum / groupItemC)
groupItemV = (groupItemP * groupItemC)

# New DF with formatting
Pop_Item = pd.DataFrame({"Purchase Count": groupItemC,
                          "Item Price": groupItemP,
                          "Total Purchase Value": groupItemV})

Pop_Item = Pop_Item.sort_values("Purchase Count", ascending=False) 
Pop_Item["Item Price"] = Pop_Item["Item Price"].map("${:.2f}".format)
Pop_Item["Total Purchase Value"] = Pop_Item["Total Purchase Value"].map("${:.2f}".format)
Pop_Item = Pop_Item[["Purchase Count", "Item Price", "Total Purchase Value"]]
Pop_Item.head()
```
